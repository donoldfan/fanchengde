```python
#统计全班的成绩
import pandas as pd
score={'语文':[68,95,98,90,80],
       '数学':[65,76,86,88,90],
       '英语':[30,98,88,77,90]}
df=pd.DataFrame(score,index=['张飞','关羽','刘备','典韦','许褚'])
df.index.name='姓名'
df.describe().loc[['mean','max','min','std']]
```

```python
df['sum']=df.sum(axis=1)
df['名次']=df['sum'].rank(ascending=False)
df
```

```python
#分析汽车投诉抱怨信息：对汽车质量数据进行统计
import pandas as pd
result=pd.read_csv('C:\python\car_complain12.csv')
result
```

```python
result=result.drop('problem',1).join(result.problem.str.get_dummies(','))
result.to_csv('car_problems.csv')
print(result.columns)
tags=result.columns[7:]
```

```python
#各品牌抱怨数量
df=result.groupby(['brand'])['id'].agg(['count'])
df=df.sort_values('count',ascending=False)
df
```

```python
#各品牌车型抱怨数量
df_model=result.groupby(['brand','car_model'])['id'].agg(['count'])
df_model=df_model.sort_values('count',ascending=False)
df_model
```

```python
#哪个品牌的平均车型投诉最多
df=result.groupby(['brand'])['id'].agg(['count'])
df2=result.groupby(['brand','car_model'])['id'].agg(['count']).groupby(['brand']).mean().sort_values('count',ascending=False)
df2
```
